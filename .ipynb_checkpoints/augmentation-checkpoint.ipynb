```python
from skimage.io import imread
from tqdm import tqdm_notebook as tqdm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import cv2
import os
```

```python
# Define params and helpers
xlsx_path = './ODIR_dataset/ODIR-5K_Training_Annotations(Updated)_V2.xlsx'
data_dir = './ODIR_dataset/ODIR-5K_Training_Dataset/'

data_dir_abs = '/Users/Natthawat/Desktop/works/research/ODIR/ODIR_dataset/'
train_data_dir_abs = '/Users/Natthawat/Desktop/works/research/ODIR/ODIR_dataset/ODIR-5K_Training_Dataset'
```

```python
df = pd.read_excel(xlsx_path)
df[:5]
```

```python
def get_input(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return(img)
```

```python
# Fix normal img paths
def relative_to_abs_path(path):
    new_path = train_data_dir_abs + '/' + path
    return new_path
```

```python
left_img_list = df['Left-Fundus'].values.tolist()
right_img_list = df['Right-Fundus'].values.tolist()
```

```python
left_temp = []
right_temp = []
for i in tqdm(range(len(left_img_list))):
    left_temp.append(relative_to_abs_path(left_img_list[i]))
    right_temp.append(relative_to_abs_path(right_img_list[i]))
    
df['Left-Fundus'] = left_temp
df['Right-Fundus'] = right_temp
```

```python
df[:5]
```

```python
aug_data_dir = './ODIR_dataset/augment_imgs/'

if not os.path.exists(aug_data_dir):
    os.makedirs(aug_data_dir)
```

```python
def flip(img, mode):
    '''
        mode: int; {0, 1}
        if mode == 0: Vertical flip
        if mode == 1: Horizontal flip
        
        img: n_array
    '''
    return cv2.flip(img, mode)
```

```python
vertical_path = os.path.join(aug_data_dir, 'vertical')

if not os.path.exists(vertical_path):
    os.makedirs(vertical_path)
```

```python
vertical_flip_df = df.copy()
left_img_list = vertical_flip_df['Left-Fundus'].values.tolist()
right_img_list = vertical_flip_df['Right-Fundus'].values.tolist()

left_img_temp = []
right_img_temp = []
for i in tqdm(range(len(left_img_list))):
    im_l = cv2.imread(left_img_list[i])
    im_r = cv2.imread(right_img_list[i])
    
    im_l_flip = flip(im_l, 0)
    im_r_flip = flip(im_r, 0)
    
    im_l_file_name = left_img_list[i].split('/')[-1]
    im_r_file_name = right_img_list[i].split('/')[-1]
    save_path_l = vertical_path + '/vertical_flip_' + im_l_file_name
    save_path_r = vertical_path + '/vertical_flip_' + im_r_file_name
    
    left_img_temp.append(save_path_l)
    right_img_temp.append(save_path_r)
    
    cv2.imwrite(save_path_l, im_l_flip)
    cv2.imwrite(save_path_r, im_r_flip)
    
vertical_flip_df['Left-Fundus'] = left_img_temp
vertical_flip_df['Right-Fundus'] = right_img_temp

vertical_df_path = data_dir_abs + '/vertical_df.csv'
vertical_flip_df.to_csv(vertical_df_path, index=False)
```

```python
horizontal_path = os.path.join(aug_data_dir, 'horizontal')

if not os.path.exists(horizontal_path):
    os.makedirs(horizontal_path)
```

```python
horizontal_flip_df = df.copy()
left_img_list = horizontal_flip_df['Left-Fundus'].values.tolist()
right_img_list = horizontal_flip_df['Right-Fundus'].values.tolist()

left_img_temp = []
right_img_temp = []
for i in tqdm(range(len(left_img_list))):
    im_l = cv2.imread(left_img_list[i])
    im_r = cv2.imread(right_img_list[i])
    
    im_l_flip = flip(im_l, 0)
    im_r_flip = flip(im_r, 0)
    
    im_l_file_name = left_img_list[i].split('/')[-1]
    im_r_file_name = right_img_list[i].split('/')[-1]
    save_path_l = horizontal_path + '/vertical_flip_' + im_l_file_name
    save_path_r = horizontal_path + '/vertical_flip_' + im_r_file_name
    
    left_img_temp.append(save_path_l)
    right_img_temp.append(save_path_r)
    
    cv2.imwrite(save_path_l, im_l_flip)
    cv2.imwrite(save_path_r, im_r_flip)
    
horizontal_flip_df['Left-Fundus'] = left_img_temp
horizontal_flip_df['Right-Fundus'] = right_img_temp

horizontal_df_path = data_dir_abs + '/horizontal_df.csv'
horizontal_flip_df.to_csv(horizontal_df_path, index=False)
```
